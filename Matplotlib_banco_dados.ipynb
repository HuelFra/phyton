```python
# Importando os pacotes
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfsalario = pd.read_csv("Salaries.csv")
```

```python
dfsalario.head(2)
```

```python
dfsalario = dfsalario[dfsalario["BasePay"] != "Not Provided"]
```

```python
dfsalario["BasePay"] = dfsalario["BasePay"].astype(float)
```

```python
dfsalario_plots = dfsalario[["BasePay","Year", "JobTitle"]]
```

```python
dfsalario_plots
```

```python

dfsalario_plots.groupby("Year").size()
```

```python
dfsalario_plots["Year"] = dfsalario_plots["Year"].astype(int)
qtd_por_ano = dfsalario_plots.groupby("Year").size()
```

```python
plt.figure(figsize=(10,5))
plt.bar(qtd_por_ano.index, qtd_por_ano.values)
plt.xlabel("Ano")
plt.ylabel("Quantidade")
plt.title("Quantidade de pessoas por ano")
plt.show()
```

```python
media_salario = dfsalario_plots.groupby("Year")["BasePay"].mean()
```

```python
plt.figure(figsize=(10,5))
plt.plot(media_salario.index, media_salario.values)
plt.xlabel("Ano")
plt.ylabel("MÃ©dia do salÃ¡rio")
plt.title("MÃ©dia do salÃ¡rio por ano")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(dfsalario_plots["BasePay"], dfsalario_plots["Year"])
ax.set_xlabel("SalÃ¡rio")
ax.set_ylabel("Ano")
```

```python
#
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
ax1.bar(qtd_por_ano.index, qtd_por_ano.values, color='green')
ax1.set_title('Quantidade de Registros por Ano')
ax1.set_xlabel('Ano')
ax1.set_ylabel('Quantidade')
ax2.plot(media_salario.index, media_salario.values, marker='o')
ax2.set_title('MÃ©dia Salarial por Ano')
ax2.set_xlabel('Ano')
ax2.set_ylabel('MÃ©dia Salarial')
plt.tight_layout()
plt.show()
```

```python
dfsalario_plots.boxplot(column='BasePay', by='Year', grid=True)
plt.title('DistribuiÃ§Ã£o dos SalÃ¡rios por Ano')
plt.xlabel('Ano')
plt.ylabel('SalÃ¡rio Base')
plt.show()
```
