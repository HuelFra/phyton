```python
import json

def adicionar_tarefa(tarefas):
    descricao = input("Descrição da tarefa: ")
    prioridade = input("Prioridade (alta, média, baixa): ").lower()
    tarefas.append({"descricao": descricao, "prioridade": prioridade})
    print("Tarefa adicionada com sucesso!")

def remover_tarefa(tarefas):
    if not tarefas:
        print("Nenhuma tarefa para remover.")
        return
    listar_tarefas(tarefas)
    try:
        indice = int(input("Índice da tarefa para remover: "))
        if 1 <= indice <= len(tarefas):
            del tarefas[indice - 1]
            print("Tarefa removida com sucesso!")
        else:
            print("Índice inválido.")
    except ValueError:
        print("Entrada inválida. Digite um número.")

def listar_tarefas(tarefas):
    if not tarefas:
        print("Nenhuma tarefa cadastrada.")
        return
    print("\nLista de Tarefas:")
    for indice, tarefa in enumerate(tarefas, 1):
        print(f"{indice}. {tarefa['descricao']} - Prioridade: {tarefa['prioridade']}")

def priorizar_tarefas(tarefas):
    if not tarefas:
        print("Nenhuma tarefa cadastrada.")
        return
    prioridade_ordem = {"alta": 1, "média": 2, "baixa": 3}
    tarefas.sort(key=lambda x: prioridade_ordem.get(x['prioridade'], 4))
    print("Tarefas ordenadas por prioridade!")

def salvar_tarefas(tarefas):
    with open('tarefas.json', 'w') as arquivo:
        json.dump(tarefas, arquivo, indent=2, ensure_ascii=False)
    print('Tarefas salvas com sucesso!')

def carregar_tarefas():
    try:
        with open("tarefas.json", "r") as arquivo:
            tarefas = json.load(arquivo)
            print("Tarefas carregadas com sucesso!")
            return tarefas
    except FileNotFoundError:
        print("Nenhum arquivo de tarefas encontrado. Iniciando uma nova lista.")
        return []

def menu():
    tarefas = carregar_tarefas()
    while True:
        print("\n--- MENU DE TAREFAS ---")
        print("1. Adicionar tarefa")
        print("2. Listar tarefas")
        print("3. Ordenar por prioridade")
        print("4. Remover tarefa")
        print("5. Salvar tarefas")
        print("6. Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            adicionar_tarefa(tarefas)
        elif opcao == "2":
            listar_tarefas(tarefas)
        elif opcao == "3":
            priorizar_tarefas(tarefas)
        elif opcao == "4":
            remover_tarefa(tarefas)
        elif opcao == "5":
            salvar_tarefas(tarefas)
        elif opcao == "6":
            salvar_tarefas(tarefas)
            print("Saindo... Tarefas salvas.")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Executa o menu
menu()
```
