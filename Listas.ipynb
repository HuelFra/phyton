```python
!pip install colorama
from colorama import Fore, Back, Style
minha_lista = [3,6,9]
print('A minha_lista é: ',minha_lista)

lista_vazia = []
print('A lista_vazia é: ',lista_vazia)

lista_variavel = [0, 'abacate', 4, True]
print('A lista_variavel é: ',lista_variavel)

minha_lista[0]
print('minha_lista[0] é: ',minha_lista[0])
minha_lista[1]
print('minha_lista[1] é: ',minha_lista[1])
minha_lista[2]
print('minha_lista[2] é: ',minha_lista[2])
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
lista_variavel [2]
print('lista_variavel[2] é: ',lista_variavel[2])
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
soma_listas = (minha_lista[2] + lista_variavel[2])
print('A soma da minha_lista [2] + lista_variavel[2] é: ',soma_listas)
multiplicação_listas = (minha_lista[2] * lista_variavel[2])
print('A multiplicação da minha_lista [2] * lista_variavel[2] é: ',multiplicação_listas)
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
nova_lista = [5,4,29,1,-2]
if nova_lista[1]%2 == 0:
    print(nova_lista[1] * nova_lista[2])
else:
    print(nova_lista[3] + nova_lista[4])
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
lista_nums = [1,2,3,4]
print('A lista original é:' , lista_nums)
lista_nums.append(5)
minha_lista = [3,6,9]
print('A minha_lista é: ',minha_lista)

lista_vazia = []
print('A lista_vazia é: ',lista_vazia)

lista_variavel = [0, 'abacate', 4, True]
print('A lista_variavel é: ',lista_variavel)

print('minha_lista[0] é: ',minha_lista[0])
print('minha_lista[1] é: ',minha_lista[1])
print('minha_lista[2] é: ',minha_lista[2])
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
print('lista_variavel[2] é: ',lista_variavel[2])
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
soma_listas = (minha_lista[2] + lista_variavel[2])
print('A soma da minha_lista [2] + lista_variavel[2] é: ',soma_listas)
multiplicação_listas = (minha_lista[2] * lista_variavel[2])
print('A multiplicação da minha_lista [2] * lista_variavel[2] é: ',multiplicação_listas)
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
nova_lista = [5,4,29,1,-2]
if nova_lista[1]%2 == 0:
    print(nova_lista[1] * nova_lista[2])
else:
    print(nova_lista[3] + nova_lista[4])
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
lista_nums = [1,2,3,4]
print('A lista original é:' , lista_nums)
lista_nums.append(5)
print('Com o .append(5), inserimos o número 5 no final da lista. Temos como resultado: ',lista_nums)
lista_nums.append('seis')
print('Com o .append(seis) inserimos a palavra seis no final da lista.Temos como resultado: ',lista_nums)
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
lista_nums.insert(0,'zero')
print('Com o .insert(0,zero), inserimos a palavra zero no início da lista, na posição 0. Temos como resultado: ',lista_nums)
lista_nums.insert(3,'banana')
print('Com o .insert(3,banana), inserimos a palavra banana na posição 3 da lista. Temos como resultado: ',lista_nums)
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
lista_nums.remove(1)
print('Com o .remove(1), removemos o elemento que está na posição 1. Temos como resultado: ',lista_nums)
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
print(Back.LIGHTBLUE_EX + Fore.LIGHTYELLOW_EX + " ======   Inserindo dois elementos iguais na lista, para entendermos que o que é removido é a posição e não o número   =======" + Style.RESET_ALL)
lista_nums.insert(1,1)
lista_nums.insert(6,1)
print('Com o .insert(1,1) e .insert(6,1),inserimos o número 1 na primeira posição e na sexta posição.')
print('Temos como resultado: ',lista_nums)
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
print(Back.LIGHTBLUE_EX + Fore.LIGHTYELLOW_EX + " ======   Removeremos o número 1 da primeira posição apenas   =======" + Style.RESET_ALL)
lista_nums.remove(1)
print('Removemos o número 1 da primeira posição. teremos como resultado: ',lista_nums)
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
print(Back.LIGHTBLUE_EX + Fore.LIGHTYELLOW_EX + " ======   Utilizaremos o pop, que remove o último elemento e mostra o que foi removido   =======" + Style.RESET_ALL)
lista_nums.pop() # Ao utilizar o print, o retorno do pop não aparece
print('Com o .pop(), removemos o último elemento da lista. Temos como resultado: ',lista_nums)
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
print(Back.LIGHTBLUE_EX + Fore.LIGHTYELLOW_EX + " ======  Utilizaremos o clear para limpar toda a lista   =======" + Style.RESET_ALL)
lista_nums.clear()
print('Com o .clear(), removemos todos os elementos da lista. Temos como resultado: ',lista_nums)
print(Back.GREEN +"+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++"+ Style.RESET_ALL)
```
