```python
# 1. Opera√ß√µes B√°sicas
# Escreva um programa que receba dois n√∫meros do usu√°rio e
# exiba o resultado das opera√ß√µes:
# Soma (+)
# Subtra√ß√£o (-)
# Multiplica√ß√£o (*)
# Divis√£o (/)
# Divis√£o inteira (//)
# M√≥dulo (%)
# Exponencia√ß√£o (**)
n1 =  float(input("Digite o primeiro n√∫mero: "))
n2 =  float(input("Digite o segundo n√∫mero: "))
print("***************************************************************")
print ("A soma de n1 + n2 √©: ",int(n1 + n2))
print ("A subtra√ß√£o de n1 - n2 √©: ",int (n1 - n2))
print ("A multiplica√ß√£o de n1 * n2 √©: ",int(n1 * n2))
print ("A divis√£o de n1/n2 √©: ", n1 / n2)
print ("A divis√£o interia de n1 // n2 √©: ",int(n1 // n2))
print ("O m√≥dulo de n1 # n2 √©: ",int(n1 % n2))
print ("A exponencia√ß√£o de n1 ** n2 √©: ",int(n1 ** n2))
print("***************************************************************")
print("======================= Exerc√≠cio 2 ==========================================================")
# 2. M√©dia de Notas
# Pe√ßa ao usu√°rio tr√™s notas e calcule a m√©dia delas.
# üìå F√≥rmula:
# M√©dia = nota1+nota2+nota3 / 3
nota1 = float(input("Digite a primeira nota: ",))
nota2 = float(input("Digite a segunda nota: ",))
nota3 = float(input("Digite a terceira nota: ",))
M√©dia = (nota1+nota2+nota3)/3
print(" A m√©dia calculada foi: ",M√©dia)
print("***************************************************************")
print("======================= Exerc√≠cio 3 ==========================================================")
# Crie um programa que pe√ßa o raio de um c√≠rculo e calcule
# sua √°rea.
# üìå F√≥rmula: √°rea = œÄ√óraio2
# üìå Dica: Use import math e math.pi.
import math
circ = float(input("Informe o raio de um c√≠rculo: "))
area = math.pi*(circ**2)
print ("A √°rea do c√≠rculo calculada √© de: ",format(area, ".2f"))
print("***************************************************************")
print("======================= Exerc√≠cio 4 ==========================================================")
# 4. Convers√£o de Temperatura
# Crie um programa que converta graus Celsius para
# Fahrenheit.
# F√≥rmula: (C x 9/5) +32
Cels = (float(input("Digite a temperatura em Celsius: ")))
Far = (Cels * 9/5) + 32
print( "A temperatura de ",Cels,"¬∫C √© igual a ",Far,"¬∫F")
print("***************************************************************")
print("======================= Exerc√≠cio 5 ==========================================================")
# 5. Calculadora de Desconto
# Crie um programa que pe√ßa o pre√ßo de um produto e o percentual de desconto aplicado.
# O programa deve calcular o pre√ßo final do produto
# A f√≥rmula √©:
# Pre√ßo final = Pre√ßo original - (Pre√ßo Original x (Desconto/100))
preco = float(input("Informe o pre√ßo do produto: "))
desconto = float(input("Informe o valor do desconto: "))
preco_final = preco -(preco*(desconto/100))
print("O preco final do produto, com desconte √©: ", format(preco_final, ".2f"))
print("***************************************************************")
print("======================= Exerc√≠cio 6 ==========================================================")
# 6. Maior de Idade?
# Pe√ßa ao usu√°rio sua idade e verifique se ele √© maior
# de idade (18 anos ou mais).
idade = int(input("Digite sua idade: "))
if idade >= 18:
    print("Voc√™ √© maior de idade.")
    print("***************************************************************")
else:
    print("Voc√™ √© menor de idade.")
    print("***************************************************************")
print("======================= Exerc√≠cio 7 ==========================================================")
# 7. Pode Acessar o Sistema?
# Crie um programa que verifique se um usu√°rio pode
# acessar um sistema. O acesso s√≥ √© permitido se:
# ‚úÖ Ele for administrador OU tiver mais de 18 anos E
# tiver uma conta ativa

admin = bool(input("Voc√™ √© administrador? (Sim/N√£o): "))
idade = int(input("Digite sua idade: "))
conta_ativa = bool(input("Sua conta est√° ativa? (Sim/N√£o): "))
if admin == True or (idade >= 18 and conta_ativa == True):
    print("Voc√™ pode acessar o sistema.")
else:
    print("Voc√™ n√£o pode acessar o sistema.")
print("***************************************************************")
print("======================= Exerc√≠cio 8 ==========================================================")
# 8. N√∫mero Par e Maior que 10?
# Pe√ßa ao usu√°rio um n√∫mero e verifique se ele √© par e maior
# que 10 ao mesmo tempo.
# üìå Dica: Um n√∫mero √© par quando numero % 2 == 0.
x = int(input("Digite um n√∫mero inteiro: "))

x_par = x % 2 == 0
x_maior_que_10 = x > 10

# Relat√≥rio completo
if x_par:
    print("O n√∫mero √© par.")
else:
    print("O n√∫mero n√£o √© par.")

if x_maior_que_10:
    print("O n√∫mero √© maior que 10.")
else:
    print("O n√∫mero n√£o √© maior que 10.")
print("***************************************************************")
print("======================= Exerc√≠cio 9 ==========================================================")
# 9. Pode Tirar Carteira de Motorista?
# O usu√°rio pode tirar a carteira de motorista se:
# ‚úÖ Tiver pelo menos 18 anos E n√£o tiver problemas na documenta√ß√£o.
# üìå Entrada: Perguntar a idade e se os documentos est√£o em dia (True ou False)
idade_nova = int(input("Informe sua idade: "))
documento = input(("Seus documentos est√£o em dia?"))
if idade_nova >= 18 and documento == True:
    print("Voc√™ pode tirar a carteira de motorista.")
else:
    print("Voc√™ n√£o pode tirar a carteira de motorista.")
print("***************************************************************")
print("======================= Exerc√≠cio 10 ==========================================================")
# 10. Promo√ß√£o de Loja
# Uma loja est√° fazendo uma promo√ß√£o onde os clientes ganham um brinde se:
# ‚úÖ Gastarem mais de R$100 OU forem clientes VIP (True ou False)
# üìå Entrada: Perguntar o valor gasto e se √© cliente VIP.

vip = input("Voc√™ √© cliente VIP? (Sim/N√£o): ").strip().lower()
compra = float(input("Informe quanto voc√™ gastou: "))

if (compra >= 100) or (vip == "sim"):
    print("Voc√™ ganhou um brinde.")
else:
    print("Voc√™ n√£o ganhou nenhum brinde.")

print("***************************************************************")


```
