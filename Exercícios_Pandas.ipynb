```python
# 1) Importar o pacote pandas
import numpy as np
import pandas as pd
```

```python
# 6) Qual é a base salarial média (BasePay)?

# Converter a coluna "BasePay" para numérica, substituindo strings não-numéricas por NaN
dfsalario["BasePay"] = dfsalario["BasePay"].replace("não fornecido", np.nan)
# Calcular a média ignorando os NaN
mediasalarial = dfsalario["BasePay"] = pd.to_numeric(dfsalario["BasePay"], errors='coerce')

print(mediasalarial)
```

```python
# 7) Qual é o maior OvertimePay do dataset?

#Converter a coluna para o tipo numérico
dfsalario["OvertimePay"] = dfsalario["OvertimePay"].replace("não fornecido", np.nan)
dfsalario["OvertimePay"] = pd.to_numeric(dfsalario["OvertimePay"])

#Calculando o valor máximo de OvertimePay
dfsalario["OvertimePay"].max()
```

```python
# 8) Qual é o cargo de JOSEPH DRISCOLL?
dfsalario["JobTitle"] = dfsalario["JobTitle"].replace("não fornecido", np.nan)
dfsalario[dfsalario["EmployeeName"] == "JOSEPH DRISCOLL"]["JobTitle"]
```

```python
# 9) Quanto ele ganha, incluindo banefícios?
dfsalario["TotalPay"] = dfsalario["TotalPay"].replace("não fornecido", np.nan)
dfsalario[dfsalario["EmployeeName"]== "JOSEPH DRISCOLL"]["TotalPay"]
```

```python
# 10) Quem é a pessoa que tem o maior salário base?
dfsalario["BasePay"] = dfsalario["BasePay"].replace("não fornecido", np.nan)
dfsalario[dfsalario["BasePay"] == dfsalario["BasePay"].max()]
```

```python
# 11) Quem é a pessoa mais mal paga?
dfsalario["BasePay"] = dfsalario["BasePay"].replace("não fornecido", np.nan)
dfsalario[dfsalario["BasePay"] == dfsalario["BasePay"].min()]
```

```python
# 12) Qual a média salarial por ano?
dfsalario.groupby("Year")["BasePay"].mean()
```
